# Retrieve Original Dataset

```python
import pandas as pd
import os

temperature_data = pd.read_csv(os.path.join("historical-hourly-weather-data/temperature.csv"))
humidity_data = pd.read_csv(os.path.join("historical-hourly-weather-data/humidity.csv"))
pressure_data = pd.read_csv(os.path.join("historical-hourly-weather-data/pressure.csv"))
windspeed_data = pd.read_csv(os.path.join("historical-hourly-weather-data/wind_speed.csv"))
locations_data = pd.read_csv(os.path.join("historical-hourly-weather-data/city_attributes.csv"))

```

### Identify Null Subsets

```python
# Extract all null rows from original dataset
null_temps = temperature_data[temperature_data.isnull().any(axis=1)]
years = ['2012', '2013', '2014', '2015', '2016', '2017']

# Iterate over every year and count how many nulls exist for each subset year
for year in years:
    nulls_year = null_temps[(temperature_data['datetime'] >= (year + '-01-01')) & (temperature_data['datetime'] <= (year + '-12-31'))]
    print(year, 'nulls --', nulls_year.size)
```

# Generate & Store Customized Data Subsets

```python
select_cities = ['Seattle', 'Los Angeles', 'Chicago', 'Miami', 'New York']

# Create dataframes with pertinent weather data (temperature, humidity, pressure) for each selected city
for city in select_cities:
    weather_df = pd.DataFrame({'datetime': temperature_data['datetime'], 
                               'temperature': temperature_data[city], 
                               'humidity': humidity_data[city], 
                               'pressure': pressure_data[city], 
                               'wind_speed': windspeed_data[city]})
    
    # Use the 2016 data subset, since it's the most recent year subset with no nulls
    weather_df = weather_df[(weather_df['datetime'] >= '2016-01-01') & (weather_df['datetime'] <= '2016-12-31')]
    weather_df.dropna()
    
    # Convert the temperatures from Kelvin to Fahrenheit
    weather_df['temperature'] = ((weather_df['temperature'] - 273.15) * 1.8) + 32
    weather_df['temperature'] = weather_df['temperature'].astype(int)
    
    # Write the dataframes out to files
    weather_df.to_csv(path_or_buf = 'city-specific-data/' + city.lower() + '_weather.csv', index=False)
```

### Retrieve Customized Subsets

# Question 1:  What are the cities' mean temperatures?

### Mean Annual Temperature

```python
import numpy as np

# Store calculated averages
avg_temps = []

# Calculate & display annual averages for each city
for city in national_weather:
    avg_annual_temp = np.mean(national_weather[city]['temperature'])
    print(city, '--', '{:.1f}Â°F'.format(avg_annual_temp))
    avg_temps.append(avg_annual_temp)
```

### Mean Seasonal Temperature

```python
for city in national_weather:
    print(city)
    
    # Retrieve year-round weather data for each city
    city_weather = national_weather[city]
    
    # Section city's year-round weather into seasons
    spring_weather = city_weather[(city_weather['datetime'] >= '2016-03-21') & (city_weather['datetime'] < '2016-06-21')]
    summer_weather = city_weather[(city_weather['datetime'] >= '2016-06-21') & (city_weather['datetime'] < '2016-09-21')]
    fall_weather = city_weather[(city_weather['datetime'] >= '2016-09-21') & (city_weather['datetime'] < '2016-12-21')]
    winter_weather = city_weather[(city_weather['datetime'] >= '2016-12-21') | (city_weather['datetime'] < '2016-03-22')]

    # Display mean seasonal temperatures
    print('Spring -- {:.1f}Â°F'.format(np.mean(spring_weather['temperature'])))
    print('Summer -- {:.1f}Â°F'.format(np.mean(summer_weather['temperature'])))
    print('Fall -- {:.1f}Â°F'.format(np.mean(fall_weather['temperature'])))
    print('Winter -- {:.1f}Â°F'.format(np.mean(winter_weather['temperature'])))

    print('\n')
```

# Question 2: How does temperature change over the course of a day?

```python
import matplotlib.pyplot as plt

for city in national_weather:
    
    # Retrieve year-round weather data for each city
    city_weather = national_weather[city]
    
    # Selecting the data for the starting days for each seasons
    season_starts = city_weather[((city_weather['datetime'] >= '2016-03-21 00:00:00') & (city_weather['datetime'] <= '2016-03-21 23:00:00')) | 
                                     ((city_weather['datetime'] >= '2016-06-21 00:00:00') & (city_weather['datetime'] <= '2016-06-21 23:00:00')) |
                                     ((city_weather['datetime'] >= '2016-09-21 00:00:00') & (city_weather['datetime'] <= '2016-09-21 23:00:00')) |
                                     ((city_weather['datetime'] >= '2016-12-21 00:00:00') & (city_weather['datetime'] <= '2016-12-21 23:00:00'))]
    
    season_starts['day'] = pd.to_datetime(season_starts['datetime']).apply(lambda x: '{year}-{month}-{day}'.format(year=x.year, month=x.month, day=x.day))
    
    # Plot the change in temperature for each day from the season_start_days dataframe
    x, y = 'datetime', 'temperature'
    fig, ax = plt.subplots()
    for name, df in season_starts.groupby('day'):
        ax.scatter(x=x, y=y, data=df, label=name)
    
    ax.set_title(city)
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.legend();
```

# Question 3: How does a city's location affect its temperature?

```python
# Create a dataframe that stores the positional coordinates & mean temperatures of the select cities
select_locations_data = locations_data[locations_data['City'].isin(select_cities)]
locations_df = pd.DataFrame({'city': select_cities, 
                             'latitude': select_locations_data['Latitude'], 
                             'longitude': select_locations_data['Longitude'],
                             'mean_temp': avg_temps})

# Write the dataframe to a file
locations_df.to_csv(path_or_buf = 'city-specific-data/city_locations.csv', index=False)
```
